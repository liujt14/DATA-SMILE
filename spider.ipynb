```python
#random 


def print_main_page(movie_name):
    main_page = "main_page.html"
    with open(os.path.join(movie_name,main_page),'w') as f:
        f.write(driver.page_source)
        f.close()
    more_comments = wait.until(EC.element_to_be_clickable((By.PARTIAL_LINK_TEXT,"更多短评")))
    more_comments.send_keys('\n')
#go back to home page

def search_movie_print_home_page(movie_name):
    
    #make dir
    directory_name = movie_name
    os.makedirs(directory_name)
    
    search_text = driver.find_element_by_id("inp-query")
    search_text.send_keys(movie_name)
    search_text.submit()
    
    details = wait.until(EC.visibility_of_element_located((By.PARTIAL_LINK_TEXT,movie_name+" ")))
    details.send_keys('\n')
    #ensure page loaded
    movie_description = wait.until(EC.visibility_of_element_located((By.CLASS_NAME,"article")))
    
    main_page = "main_page.html"
    with open(os.path.join(directory_name,main_page),'w') as f:
        f.write(driver.page_source)
        f.close()
    more_comments = wait.until(EC.element_to_be_clickable((By.PARTIAL_LINK_TEXT,"更多短评")))
    more_comments.send_keys('\n')

def write_files_after_loaded(movie_name,file_name):
    i = 1
    next_page = wait.until(EC.visibility_of_element_located((By.CLASS_NAME,"next")))
    random_time = random.randint(50, 100)/100

    time.sleep(random_time)
    #25_pages
    while True:
    #file write into html
        file_names = file_name+str(i)+".html"
        print(file_names)
        with open(os.path.join(movie_name,file_names),'w') as f:
            f.write(driver.page_source)
            f.close()
        # go to next page
        next_page = wait.until(EC.visibility_of_element_located((By.CLASS_NAME,"next")))
        herf_data = next_page.get_attribute('href')
        if herf_data is None:
            break
        next_page.send_keys('\n')
        previous_page = wait.until(EC.visibility_of_element_located((By.LINK_TEXT,"< 前页")))
        i += 1 


def print_pages(movie_name):
    #step1_print all reviews
    write_files_after_loaded(movie_name,"all_reviews") 
    #step2_print positive reviews
    input_h = wait.until(EC.visibility_of_element_located((By.XPATH,"//input[@value='h']")))
    input_h.click()
    write_files_after_loaded(movie_name,"positive_reviews")
    #step3_print middle reviews
    input_m = wait.until(EC.visibility_of_element_located((By.XPATH,"//input[@value='m']")))
    input_m.click()
    write_files_after_loaded(movie_name,"middle_reviews")
    #step4_print negative reviews
    input_l = wait.until(EC.visibility_of_element_located((By.XPATH,"//input[@value='l']")))
    input_l.click()
    write_files_after_loaded(movie_name,"negative_reviews")
    
    #step3_print lastest reviews
    lastest_reviews = wait.until(EC.element_to_be_clickable((By.LINK_TEXT,"最新")))
    lastest_reviews.send_keys('\n')
    
    write_files_after_loaded(movie_name,"lastest_reviews")
```
